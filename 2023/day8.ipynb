```python
def getlines(num="8"):
    with open(f"inputs/day{num}.txt") as f: lines = f.readlines()
    lines = [line.strip() for line in lines]
    return lines 
# lines
```

```python
lrs = lines[0]
lrs
```

```python
def extract(line):
    start, leftright = line.split(" = ")
    left, right = leftright.split(", ")
    left, right = left[1:], right[:-1]
    return start, left, right
```

```python
def graph(lines):
    g = {}
    for line in lines[2:]:
        start, left, right = extract(line)
        g[start] = (left, right)
    return g
```

```python
g
```

```python
pos = "AAA"
steps = 0
while True:
    for lr in lrs:
        pos = (g[pos][0] if lr == "L" else g[pos][1])
        steps += 1
        # print(pos)
    if pos == "ZZZ":
        break
steps
```
