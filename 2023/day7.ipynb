```python
with open("inputs/day7.txt") as f: lines = f.readlines()
lines = [line.strip() for line in lines]
lines
```

```python
cards = list('23456789TJQKA')
def cardrank(c): return cards.index(c) + 2
cardrank("J")
```

```python
def gettype(hand):
    s = set(hand)
    # 5 of a kind
    if len(s) == 1:
        return 7
    if len(s) == 5:
        return 1
    counts = sorted(Counter(hand).values())
    # 4 or a kind
    if counts == [1,4]:
        return 6
    # Full house
    if counts == [2,3]:
        return 5
    # 3 of a kind
    if counts == [1,1,3]:
        return 4
    # 2 pairs
    if counts == [1,2,2]:
        return 3
    if counts == [1,1,1,2]:
        return 2
    

gettype("KKKKK"), gettype("JJJJ2"), gettype("22333"), gettype("23555"), gettype("A2345")
```

```python
"A23JJ".replace("J","K")
```

```python
def compare(x1, x2):
    (hand1, bid1), (hand2, bid2) = x1, x2
    t1, t2 = gettype(hand1), gettype(hand2)
    if t1 == t2:
        for c1, c2 in zip(hand1, hand2):
            if cardrank(c1) == cardrank(c2):
                continue
            if cardrank(c1) > cardrank(c2):
                return 1
            else:
                return -1
    return 1 if t1 > t2 else -1    
```

```python
from functools import cmp_to_key
hands = [line.split(" ") for line in lines]
sorted_hands = sorted(hands, key=cmp_to_key(compare))
sorted_hands
```

```python
sum = 0
for i , (hand, bid) in enumerate(sorted_hands):
    rank = i + 1
    sum += rank * int(bid)
sum
```

```python
newcards = list('J23456789TQKA')
def newrank(c): return newcards.index(c) + 2
newrank("J")
```

```python
def newtype(hand):
    njacks = len([x for x in hand if x == "J"])
    if njacks == 0:
        return gettype(hand)
    if njacks == 5:
        return 7 
    return max([gettype(hand.replace("J", x)) for x in newcards[1:]])
```

```python
def newcompare(x1, x2):
    (hand1, bid1), (hand2, bid2) = x1, x2
    t1, t2 = newtype(hand1), newtype(hand2)
    if t1 == t2:
        for c1, c2 in zip(hand1, hand2):
            if newrank(c1) == newrank(c2):
                continue
            if newrank(c1) > newrank(c2):
                return 1
            else:
                return -1
    return 1 if t1 > t2 else -1    
```

```python
sorted_hands = sorted(hands, key=cmp_to_key(newcompare))
sum = 0
for i , (hand, bid) in enumerate(sorted_hands):
    rank = i + 1
    sum += rank * int(bid)
sum
```
